```python
fig , (ax1,ax2,ax3,ax4) = plt.subplots(4,1 , figsize = (6,8))

axes = [ax1,ax2,ax3,ax4]

for ax in axes:
    
    ax.plot(death_tolls["Month"] , death_tolls["New_deaths"] , color = "#f24162" , alpha = 0.4)
    
    # removing labels
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    
    # removing params/sticks
    ax.tick_params(bottom = 0 , left = 0)
    
    # removing borders/spines
    for location in ["left" , "right" , "top" , "bottom"]:
        ax.spines[location].set_visible(False) 

# plotting quaterly months with dark color
ax1.plot(death_tolls["Month"][:3] , death_tolls["New_deaths"][:3] , color = "#f20f2d" , linewidth = 2.6)
ax2.plot(death_tolls["Month"][2:6] , death_tolls["New_deaths"][2:6] , color = "#f20f2d" , linewidth = 2.6)
ax3.plot(death_tolls["Month"][5:9] , death_tolls["New_deaths"][5:9] , color = "#f20f2d" , linewidth = 2.6)
ax4.plot(death_tolls["Month"][8:12] , death_tolls["New_deaths"][8:12] , color = "#f20f2d" , linewidth = 2.6)

# label the cases

# Jan - March
ax1.text(0.5, -80,"0", alpha = 0.5) # intially cases = 0
ax1.text(2.5 , 300 , "77" , alpha = 0.5) # at march cases = 77
# April - June
ax2.text(2.5, 300 ,"77", alpha = 0.5) 
ax2.text(5.9 , 1000 , "785" , alpha = 0.5) 
# July - Sept
ax3.text(5.9 , 1000 , "785" , alpha = 0.5)
ax3.text(8.8 , 1000 , "739" , alpha = 0.5)
# Oct - December
ax4.text(8.8 , 1000 , "739" , alpha = 0.5)
ax4.text(11 , 2500 , "2247" , alpha = 0.5)

# label months
ax1.text(1.1 , -300 , "Jan-Mar" , color = "#f20f2d" , weight = "bold" , rotation = 3)
ax2.text(3.4 , 200 , "Apr-Jun" , color = "#f20f2d" , weight = "bold" , rotation = 70)
ax3.text(7 , 500 , "Jul-Sep" , color = "#f20f2d" , weight = "bold" , rotation = 0)
ax4.text(10.5 , 600 , "Oct-Dec" , color = "#f20f2d" , weight = "bold" , rotation = 45)
```
